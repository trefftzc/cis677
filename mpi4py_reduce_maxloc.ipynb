```python
%%writefile mpi_max_loc.py
import sys
import time
import numpy as np
from mpi4py import MPI
import random

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
number_nodes = comm.Get_size()

# This array contains:
# A candidate value
# The rank of the node that contains that value
l = [random.random(),rank]
array = np.array(l)
print("Node: ",rank," contains: ",l)
# A reduce function with the MAXLOC operations
# expects as a parameter an array with two entries in all nodes
# The maximum value on the first entry of all those arrays is returned
# in the root node in the first entry (index 0)
# The value in the second entry of all those arrays is returend
# in the root node in the second entry (index 1)

result = comm.reduce(array,op=MPI.MAXLOC,root=0)
if (rank == 0):
  print("the max value is: ",result[0])
  print("the node that contains that max value is: ",result[1])
```
