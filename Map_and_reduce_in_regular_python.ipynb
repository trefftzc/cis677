# Map and Reduce in regular python

In regular python it is possible to declare functions that will receive as parameters other functions.

In the example the function call_another_function receives as a parameter a list and another function.
It applies the parameter function to all the elements in the list and returns a new list.

```python
def call_another_function(list, function):
  result = []

  for element in list:
    result.append(function(element))

  return result

```

Now, let's define two different functions.


```python
def double(a):
  return a+a

def square(b):
  return b*b
```

Now, let's invoke call_another_function twice, once with each of the functions defined above:

```python
list1 = [1,2,3,4]
list2 = call_another_function(list1, double)
list3 = call_another_function(list1, square)
print(list1)
print(list2)
print(list3)
```

# The map function in python

The map function works very similarly to the call_another_function above.

The first parameter is the function to be applied to all the elements of the second parameter.

The second parameter should be an iterable object.


```python
list1 = [1,2,3,4]
list2 = map(double,list1 )
list3 = map(square,list1)
print(list1)
for e in list2:
  print(e,end=" ")
print()
for e in list3:
  print(e,end=" ")
print()
```

# The reduce function in python

In the functools package there is reduce function that works like the other reduction operations we have seen before.

Here, lambda functions are used to describe the operation that will be carried across all the elements in the iterable collection.

Recall that the function should be commutative and associative.



```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Summing all numbers in the list
sum_result = reduce(lambda x, y: x + y, numbers)
print(f"Sum: {sum_result}")  # Output: Sum: 15

# Calculating the product of all numbers
product_result = reduce(lambda x, y: x * y, numbers)
print(f"Product: {product_result}") # Output: Product: 120
```
