```python
import findspark
from pyspark.sql import SparkSession

# Initialize findspark to set SPARK_HOME and add PySpark to sys.path
findspark.init()

# Create SparkSession
spark = SparkSession.builder.appName("Example").getOrCreate()

print("Spark session created successfully!")
```

```python
import numpy as np
x = np.linspace(-4,4,1000)
#print(x)
```

```python
x_rdd = spark.sparkContext.parallelize(x)
```

```python
print(x_rdd)
```

The definition of the function

```python
def f(x):
    return x**4 - 10*x**2 + 9
```

```python
print (f(-4.0),f(-2.2),f(-1.0),f(0.0),f(1.0),f(2.2),f(4.0))
```

```python
result_rdd = x_rdd.map(f)
```

```python
print(result_rdd)
```

```python

max_value = result_rdd.reduce(max)
min_value = result_rdd.reduce(min)
print("The min value is: ",min_value)
print("The max value is: ",max_value)
```
